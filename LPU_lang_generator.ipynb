```python
text = text.lower()
print(text)
```

```python
T = {}
k = 4
for i in range(len(text)-k):
    X = text[i:i+k]
    Y = text[i+k]
    # print(X,Y)

    if T.get(X) is None:
        T[X] = {}
        T[X][Y] = 1
    else:
        if T[X].get(Y) is None:
            T[X][Y] = 1
        else:
            T[X][Y] += 1
```

```python
# Convert freq into probability
```

```python
for kx in T.keys():
    s = sum(T[kx].values())
    for ky in T[kx].keys():
        T[kx][ky] = T[kx][ky]/s


print(T)
```

```python
def predNextChar(sentence):
    state = sentence[-4:]

    if(T.get(state) is None):
        return " "

    possible_keys = list(T[state].keys())
    possible_values = list(T[state].values())
    return np.random.choice(possible_keys,p=possible_values)

```

```python
predNextChar("कुछ वर्षों म")
```

```python
output = "कुछ वर्षों म"
for i in range(500):
    next_char = predNextChar(output)
    output = output + next_char

print(output)
```
